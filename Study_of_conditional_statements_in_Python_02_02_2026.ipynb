```python
# Get input from user for 5 subjects and find average
# Calculate grade using if-elif-else
subject_1=int(input("Enter the grade of subject 1:"))
subject_2=int(input("Enter the grade of subject 2:"))
subject_3=int(input("Enter the grade of subject 3:"))
subject_4=int(input("Enter the grade of subject 4:"))
subject_5=int(input("Enter the grade of subject 5:"))
average=(subject_1+subject_2+subject_3+subject_4+subject_5)/5
print("The average grade is:",average)
if(average>95):
  print("O grade obtained")
elif(average>85):
  print("A grade obtained")
elif(average>75):
  print("B grade obtained")
elif(average>65):
  print("C grade obtained")
else:
  print("Fail")
```

```python
'Write a python program to calculate the gross salary of an employee based on basic salary'
'If basic <= 10000: HRA = 20%,DA = 80%'
'If basic <= 20000: HRA = 25%,DA = 90%'
'Otherwise,HRA = 30%,DA = 95%'
basic=int(input("Enter the current salary"))
if basic <= 10000:
         hra = 0.20 * basic
         da = 0.80 * basic
elif basic <= 20000:
         hra = 0.25 * basic
         da = 0.90 * basic
else:
         hra = 0.30 * basic
         da = 0.95 * basic

gross_salary = basic + hra + da
print("The gross salary is:",gross_salary)
```
